```python
from sklearn.metrics import r2_score
def score(X_train, X_test, y_train, y_test,model):
    train_predict = model.predict(X_train)
    test_predict = model.predict(X_test)
    print(f"training r2 score - {np.round(r2_score(y_train, train_predict),2)}")
    print(f"test r2 score - {np.round(r2_score(y_test, test_predict),2)}")
```

```python
#training the neural model
tf.random.set_seed(42)

model_tf = tf.keras.Sequential([
  tf.keras.layers.Dense(10,activation="relu"),
  tf.keras.layers.Dense(5,activation="relu"),
  tf.keras.layers.Dense(3,activation="relu"),
  tf.keras.layers.Dense(1)
])


model_tf.compile(loss=tf.keras.losses.mse,
              optimizer=tf.keras.optimizers.Adam(),
              metrics=["accuracy"])

# Fit the model
model_tf.fit(X_train, y_train,epochs=20,batch_size=1000,validation_split=0.2)
```
